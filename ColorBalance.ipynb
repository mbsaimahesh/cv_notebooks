```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
im = cv2.imread('/home/sravya/Documents/Code/Data/standard_test_images/lena_color_256.tif',cv2.IMREAD_COLOR)
im = cv2.cvtColor(im,cv2.COLOR_BGR2RGB)
color_trans = np.eye(3)*0.25
out = np.zeros(im.shape,dtype=np.uint8)
```

```python
print(color_trans)
```

```python
for i in range(0,256):
    for j in range(0,256):
        out[i,j,:]=np.transpose(np.matmul(color_trans,(im[i,j,:].reshape(3,1))))
        
```
